```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# ==========================================
# PART 1: EMAIL SPAM DETECTION (NLP)
# ==========================================
print("--- Training Email Spam Model ---")
# Load your email dataset
# Assuming columns: 'raw_text' (email content) and 'label' (0 for ham, 1 for spam)
df_email = pd.read_csv(r'C:\Users\SREEVIBU\Downloads\spam\email_dataset_100k.csv')

cv = CountVectorizer()
X_email = cv.fit_transform(df_email['raw_text'])
y_email = df_email['label']

x_train_e, x_test_e, y_train_e, y_test_e = train_test_split(X_email, y_email, test_size=0.2)
email_model = MultinomialNB()
email_model.fit(x_train_e, y_train_e)
print(f"Email Model Accuracy: {accuracy_score(y_test_e, email_model.predict(x_test_e))*100:.2f}%")

# ==========================================
# PART 3: TEST IT LIVE
# ==========================================
def check_threat(user_input, mode="email"):
    if mode == "email":
        data = cv.transform([user_input])
        prediction = email_model.predict(data)
        return "SPAM 🚨" if prediction[0] == 1 else "SAFE ✅"
    else:

        return "URL detection is currently disabled due to missing data."

print("\n--- Live Test ---")
print("Email Test:", check_threat("Winner! Claim your free gift card now!", mode="email"))
```
